```python
from facebook_scraper import get_posts
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
post_df_full = pd.DataFrame(columns = [])
```

```python
for post in get_posts('cnninternational', cookies='cookies.txt', extra_info=True, pages=7, options={"comments": True}):
    post_entry = post
    fb_post_df = pd.DataFrame.from_dict(post_entry, orient='index')
    fb_post_df = fb_post_df.transpose()
    post_df_full = post_df_full.append(fb_post_df)
    print(post['post_id']+' get')
```

```python
from pyspark.ml.feature import Word2Vec
```

```python
from pyspark.context import SparkContext
from pyspark.sql.session import SparkSession
sc = SparkContext('local')
spark = SparkSession(sc)
```

```python
from pyspark.sql import SQLContext
sqlContext = SQLContext(sc)
from pyspark.sql.types import StructType,StructField, StringType
```

```python
emptyRDD = spark.sparkContext.emptyRDD()
from pyspark.sql.types import *

schema = StructType([
  StructField('v1', FloatType(), True),
  StructField('v2', FloatType(), True),
  StructField('v3', FloatType(), True)
  ])

df = spark.createDataFrame(emptyRDD,schema)
df.printSchema()

df2=post_df_full[["text"]]
#df2["text"]
rows=[]
for i in df2["text"]:
  
  row=[]
  row.append(i.split(" "))
  rows.append(row)
  row=[]


columns = ['comment']
 
# Creating the DataFrame
second_df = spark.createDataFrame(rows, columns)
 
# Printing the non-empty DataFrame
second_df.show()
```

```python
#documentDF = spark.createDataFrame(df2)

# Learn a mapping from words to Vectors.
word2Vec = Word2Vec(vectorSize=3, minCount=0, inputCol="comment", outputCol="result")
model = word2Vec.fit(second_df)

result = model.transform(second_df)
for row in result.collect():
    text, vector = row

    print("Text: [%s] => \nVector: %s\n" % (", ".join(text), str(vector)))
```

```python
rows1=[]
for row in result.collect():
    text, vector = row
    rows1.append(vector.tolist())
print(rows1)
columns1 = ['v1','v2','v3']
 
# Creating the DataFrame
second_df2 = spark.createDataFrame(rows1, columns1)
```

```python

from pyspark.ml.feature import VectorAssembler
vecAssembler = VectorAssembler(inputCols=["v1","v2","v3"],outputCol="features")
new_df=vecAssembler.transform(second_df2)
new_df.show()
```

```python
from pyspark.ml.clustering import KMeans
from pyspark.ml.evaluation import ClusteringEvaluator

# Loads data.
#dataset = spark.read.format("libsvm").load("/content/datakmeans.txt")
dataset=new_df
# Trains a k-means model.
kmeans = KMeans().setK(2).setSeed(1)
model = kmeans.fit(new_df.select('features'))

# Make predictions
predictions = model.transform(new_df)

# Evaluate clustering by computing Silhouette score
evaluator = ClusteringEvaluator()

silhouette = evaluator.evaluate(predictions)
print("Silhouette with squared euclidean distance = " + str(silhouette))

# Shows the result.
centers = model.clusterCenters()
print("Cluster Centers: ")
for center in centers:
    print(center)
```
